```python
import time
from matplotlib import pyplot as plt
import numpy as np
import mxnet as mx
from mxnet import autograd, gluon
import gluoncv as gcv
from gluoncv.utils import download, viz
```

```python
path='kkk.jpg'
net = gcv.model_zoo.get_model('ssd_512_mobilenet1.0_custom', classes=['mask','no_mask'], pretrained_base=False)
net.load_parameters('best_model.params')
x, image = gcv.data.transforms.presets.ssd.load_test(path, 512)
cid, score, bbox = net(x)
ax = viz.plot_bbox(image, bbox[0], score[0], cid[0], class_names=['mask','no_mask'])
print(ax)
```
